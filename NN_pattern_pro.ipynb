reading data from feature files

```python
dataset_pl = loadtxt('/content/drive/MyDrive/pattern project/features/plastic bottles/features.csv', delimiter=',')
dataset_wt = loadtxt('/content/drive/MyDrive/pattern project/features/water bottles/features.csv', delimiter=',')
```

```python
lp=len(dataset_pl)
lw=len(dataset_wt)
###platic=0 water=1 
all_dataset = np.concatenate((dataset_pl, dataset_wt))
np.random.shuffle(all_dataset)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(all_dataset, test_size=0.3, random_state=42)
```

```python
#benfsl el features men el o/p
x_train = train[:,0:5]
y_train = train[:,5]
x_test = test[:,0:5]
y_test = test[:,5]
```

```python
model = Sequential()
model.add(Dense(12, input_shape=(5,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model.fit(x_train, y_train, epochs=150, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(x_train, y_train)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
...
# make probability predictions with the model
predictions = model.predict(x_test)
# round predictions 
rounded = [round(x_test[0]) for x_test in predictions]
...
# make class predictions with the model
predictions = (model.predict(x_test) > 0.5).astype(int)
```
